```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week02/Data/sales_predictions_2023.csv')
df.head()
df.dtypes
```

```python
fig, ax = plt.subplots()
data = df['Item_Weight']
ax.boxplot(data);
ax.set_ylabel("Item_Weight((in kg))");
ax.set_xlabel("Item_Weight")
ax.set_title("Boxplot of Item_Weight");
```

```python
fig, ax = plt.subplots()
data = df['Item_Visibility']
ax.boxplot(data);
ax.set_ylabel("Item_Visibility");
ax.set_xlabel("Item_Visibility")
ax.set_title("Boxplot of Item_Visibility");
```

```python
fig, ax = plt.subplots()
data = df['Item_MRP']
ax.boxplot(data);
ax.set_ylabel("Item_MRP");
ax.set_xlabel("Item_MRP")
ax.set_title("Boxplot of Item_MRP");
```

```python
fig, ax = plt.subplots()
data = df['Outlet_Establishment_Year']
ax.boxplot(data);
ax.set_ylabel("Outlet_Establishment_Year");
ax.set_xlabel("Outlet_Establishment_Year ")
ax.set_title("Boxplot of Outlet_Establishment_Year ");
```

```python
fig, ax = plt.subplots()
data = df['Item_Outlet_Sales']
ax.boxplot(data);
ax.set_ylabel("Item_Outlet_Sales");
ax.set_xlabel("Item_Outlet_Sales")
ax.set_title("Boxplot of Item_Outlet_Sales");
```
